```python
import pandas as pd
import numpy as np
from sklearn import svm
import seaborn as sns
import matplotlib as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#importing required libraries 
```

```python
df = pd.read_excel(r"C:\Users\Asus\Desktop\feature_set.xlsx") #Reading file to dataframe
```

```python
sns.heatmap(df.corr(), annot=True); #Correlation Matrix for feature interrelationship
```

```python
X = df.drop(['Class'], axis = 1) #Preparing Feature Data Set
y = df['Class'] #Preparing Target Class Data Set
df['Class'].value_counts() #verifying class count
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3) #Splitting Training and Testing datasets
```

```python
lr = LogisticRegression(solver='saga',multi_class='auto') #Using saga for faster multiclass processing
y_pred = lr.fit(X_train, y_train)  #fitting training data to Linear Regression model
```

```python
y_pred.score(X_test, y_test) #Model classification accuracy score
```

```python
y_pred=lr.predict(X_test) #recording predictions on test data
print(classification_report(y_test, y_pred)) #generating detailed statistical report
```
